````
from dash import Dash, html, dcc, Input, Output
import pandas as pd
import plotly.express as px

# Sample data (replace with your dataset)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015] * 3,
    'Vehicle_Type': ['Sedan']*6 + ['SUV']*6 + ['Truck']*6,
    'Sales': [30000, 32000, 25000, 28000, 35000, 34000,
              15000, 16000, 14000, 15000, 18000, 17000,
              5000, 6000, 4000, 5000, 7000, 6500],
    'Recession': [1, 1, 0, 0, 0, 0] * 3
}
df = pd.DataFrame(data)

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Automobile Sales Analytics Dashboard", style={'textAlign': 'center'}),
    html.Label("Select Vehicle Type:"),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[{'label': v, 'value': v} for v in df['Vehicle_Type'].unique()],
        value='Sedan'
    ),
    html.Label("Select Period:"),
    dcc.Dropdown(
        id='period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'Recession'},
            {'label': 'Non-Recession', 'value': 'Non-Recession'}
        ],
        value='Recession'
    ),
    html.Div(id='output-container', className='output-display'),
    dcc.Graph(id='recession-graph'),
    dcc.Graph(id='yearly-graph')
])
````

````
@app.callback(
    Output('recession-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('period-dropdown', 'value')]
)
def update_recession_graph(vehicle_type, period):
    recession_val = 1 if period == 'Recession' else 0
    filtered_df = df[(df['Vehicle_Type'] == vehicle_type) & (df['Recession'] == recession_val)]
    fig = px.line(filtered_df, x='Year', y='Sales',
                  title=f'Sales Trend for {vehicle_type} during {period}',
                  markers=True)
    fig.update_layout(xaxis_title='Year', yaxis_title='Sales')
    return fig
````

````
from dash import Dash, html, dcc, Input, Output, callback
import pandas as pd
import plotly.express as px
import plotly.io as pio # Import plotly.io for saving images

# Sample data (replace with your dataset)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015] * 3,
    'Vehicle_Type': ['Sedan']*6 + ['SUV']*6 + ['Truck']*6,
    'Sales': [30000, 32000, 25000, 28000, 35000, 34000,
              15000, 16000, 14000, 15000, 18000, 17000,
              5000, 6000, 4000, 5000, 7000, 6500],
    'Recession': [1, 1, 0, 0, 0, 0] * 3
}
df = pd.DataFrame(data)

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Automobile Sales Analytics Dashboard", style={'textAlign': 'center'}),
    html.Label("Select Vehicle Type:"),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[{'label': v, 'value': v} for v in df['Vehicle_Type'].unique()],
        value='Sedan'
    ),
    html.Label("Select Period:"),
    dcc.Dropdown(
        id='period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'Recession'},
            {'label': 'Non-Recession', 'value': 'Non-Recession'}
        ],
        value='Recession'
    ),
    html.Div(id='output-container', className='output-display'),
    dcc.Graph(id='recession-graph'),
    html.Button("Download Recession Graph as PNG", id="btn-download-recession", n_clicks=0),
    dcc.Download(id="download-recession-image"),
    dcc.Graph(id='yearly-graph'),
    html.Button("Download Yearly Graph as PNG", id="btn-download-yearly", n_clicks=0),
    dcc.Download(id="download-yearly-image")
])

@callback(
    Output('recession-graph', 'figure'),
    Input('period-dropdown', 'value')
)
def update_recession_graph(selected_period):
    filtered_df = df[df['Recession'] == (1 if selected_period == 'Recession' else 0)]
    fig = px.line(filtered_df, x='Year', y='Sales', color='Vehicle_Type',
                  title=f'Sales Trends During {selected_period}')
    return fig

@callback(
    Output('download-recession-image', 'data'),
    Input('btn-download-recession', 'n_clicks'),
    Input('period-dropdown', 'value'),
    prevent_initial_call=True
)
def download_recession_graph(n_clicks, selected_period):
    if n_clicks > 0:
        filtered_df = df[df['Recession'] == (1 if selected_period == 'Recession' else 0)]
        fig = px.line(filtered_df, x='Year', y='Sales', color='Vehicle_Type',
                      title=f'Sales Trends During {selected_period}')
        # Save the figure to a bytes object
        img_bytes = pio.to_image(fig, format='png')
        return dcc.send_bytes(img_bytes, f'recession_sales_trend_{selected_period}.png')
    return None


@callback(
    Output('yearly-graph', 'figure'),
    Input('vehicle-type-dropdown', 'value')
)
def update_yearly_graph(selected_vehicle):
    filtered_df = df[df['Vehicle_Type'] == selected_vehicle]
    fig = px.line(filtered_df, x='Year', y='Sales', title=f'{selected_vehicle} Sales Trends by Year')
    return fig

@callback(
    Output('download-yearly-image', 'data'),
    Input('btn-download-yearly', 'n_clicks'),
    Input('vehicle-type-dropdown', 'value'),
    prevent_initial_call=True
)
def download_yearly_graph(n_clicks, selected_vehicle):
    if n_clicks > 0:
        filtered_df = df[df['Vehicle_Type'] == selected_vehicle]
        fig = px.line(filtered_df, x='Year', y='Sales', title=f'{selected_vehicle} Sales Trends by Year')
        # Save the figure to a bytes object
        img_bytes = pio.to_image(fig, format='png')
        return dcc.send_bytes(img_bytes, f'{selected_vehicle}_yearly_sales_trend.png')
    return None


if __name__ == '__main__':
    app.run(debug=True, mode='inline')
````

````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````
